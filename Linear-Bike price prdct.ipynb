```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/datasets/BIKE DETAILS.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data['ex_showroom_price'] = data['ex_showroom_price'].fillna(0)
```

```python
data.head()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

for i in ['name','seller_type','owner']:
  data[i]= label_encoder.fit_transform(data[i])
  data[i].unique()


```

```python
data.head()
```

```python
y=data['ex_showroom_price']
x=data.drop(['ex_showroom_price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()

model=lr.fit(x_train,y_train)
```

```python
a=model.predict(x_test)
```
